```python
import pandas as pd
data=pd.read_excel("data1.xlsx")
```

```python
df=pd.DataFrame(data)
df
```

```python
Net_change=[]
Net_Pecent_Change=[]
Dates=[]
for i in range(len(df)-1,0,-5):
    net_change=df.loc[i,'Close']-df.loc[i-1,'Open']
    Net_change.append(net_change)
    net_percent_change=((df.loc[i-1,'Open']-df.loc[i,'Close'])/df.loc[i,'Close'])*100
    Net_Pecent_Change.append("%.2f"%net_percent_change)
    Dates.append(df.loc[i,'Date']+"-"+df.loc[i-1,'Date'])
    print("\nNet_Change & Net_%_Change between dates "+df.loc[i,'Date']+"-"+df.loc[i-1,'Date']+" is %.2f %.2f"%(net_change,net_percent_change))
```

```python
print("Representation In Tabular Format")
new_df=pd.DataFrame(list(zip(Dates,Net_change,Net_Pecent_Change)),columns=['Between Dates','Net_Change','Net_%_Change'])
new_df
```
